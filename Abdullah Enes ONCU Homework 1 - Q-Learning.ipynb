```python
## If I set pieces[1] = 1, I ignore the speed and it converges in 500 episode.
## But if I set it for example 7, it converges in 800 episode.
## On the other hand, there is no need for speed in learning I think, because position is enough for keeping cartpole on center for 400 step.

import gym
import random
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

env = gym.make('CartPole-v0')
observation_space_low = np.array([-20, -100,  -math.pi, -math.pi])
observation_space_high = np.array([20, 100, math.pi, math.pi])

pieces = [16, 1, 20, 20]

gamma = 0.99
learning_rate = 0.1
epsilon = 1

def discretize(observation):
    discrete_observation = [0, 0, 0, 0]
    
    observation[2] = (observation[2]+2*math.pi)%(2*math.pi)-math.pi
    
    for i in range(0, 4):
        if pieces[i] != 1:
            discrete_observation[i] = int(((observation[i]-observation_space_low[i])/(observation_space_high[i]-observation_space_low[i]))*pieces[i])
    for i in range(0, 4):
        discrete_observation[i] = max(0, min(discrete_observation[i], pieces[i]-1))
    return discrete_observation
```
