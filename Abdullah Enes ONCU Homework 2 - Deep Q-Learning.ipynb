```python
# I implemented with momentum and beta=0.95
# However, because sometimes it doesn't converges with momentum, I add that when our model learned something,
# I make beta = 0. 
# On the other hand, I set beta 0, because it converges faster than with momentum one.
# You can set beta = 0.90 or 0.95 for seeing momentum results.

# Note: It sometimes doesn't converge in first 100 step and I restart training from scratch if it doesn't converge.
# However, It work very good and I tried it for 400000 steps and it doesn't fall down and slide.

import gym
import random
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import clear_output

env = gym.make('CartPole-v0')

def sigmoid(x):
    if x>2 or x<-2:
        print(x)
    return 0 if x<-30 else (1 if x>30 else 1.0/(1+math.exp(-x)) )
def sigmoid_der(x):
    return sigmoid(x) * (1 - sigmoid(x))

sigmoid = np.vectorize(sigmoid)
sigmoid_der = np.vectorize(sigmoid)

w = np.random.rand(5)  # Q-table creation parameters for action 0, f0
u = np.random.rand(5)  # Q-table creation parameters for action 1, f1

def Q_table(state, action):
    if action == 0:
        return np.dot(state, w)
    else:
        return np.dot(state, u)

def arrange_state(state):
    return np.append(state,1)


ok = False

while ok == False:
    
    clear_output(wait=True)
    
    w = np.random.rand(5)  # Q-table creation parameters for action 0, f0
    u = np.random.rand(5)  # Q-table creation parameters for action 1, f1

    alpha = 0.001
    beta = 0.0
    gamma = 0.999
    epsilon = 1

    Q_values = []

    vdw = 0
    vdu = 0

    for i_episode in range(1000):
        state = env.reset()
        state = arrange_state(state)

        Q = 0

        epsilon *= 0.99

        for i_step in range(1000):

            if random.random() < epsilon:
                next_action = random.randint(0,1)
            else:
                next_action = 0 if Q_table(state, 0) > Q_table(state, 1) else 1

            next_state, reward, done, info = env.step(next_action)
            next_state = arrange_state(next_state)

            Q += reward

            max_qsa = max( Q_table(next_state, 0), Q_table(next_state, 1) )

            goal = reward + gamma * max_qsa

            # with momentum
            if next_action == 0:
                dw = state * (goal-Q_table(state, 0))
                vdw = beta * vdw + (1 - beta) * dw
                w = w + alpha * vdw

                for i in range(0,5):
                    w[i] = min(10, max(-10, w[i]))

            else:
                du = state * (goal-Q_table(state, 1))
                vdu = beta * vdu + (1 - beta) * du
                u = u + alpha * vdu

                for i in range(0,5):
                    u[i] = min(10, max(-10, u[i]))

            state = next_state

        print("Round: "+i_episode.__str__() + ", Q: "+Q.__str__())
        Q_values += [Q]

        if Q>=200:
            beta = 0

        ok = False
        if i_episode>=50:
            ok = True
            for i in range(i_episode-50, i_episode):
                if Q_values[i] < 400:
                    ok = False
                    break
        if ok:
            break

        env.close()
    
plt.plot(range(1, len(Q_values)+1), Q_values)
plt.show()
```
